```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
def distnormal(b: float, r:int):
    sum = 0
    for i in range(0,r):
        sum += np.random.uniform(-b, b)
    return 0.5 * sum
```

```python
def disttriang(b:float):
    return math.sqrt(6)/2 * (np.random.uniform(-b, b) + np.random.uniform(-b, b))
```

```python
#Calculo PDF
dist = []
dist20 = []
triang = []
for i in range (0, 10000):
    dist.append(distnormal(5.0, 12))
    dist20.append(distnormal(5.0, 20))
    triang.append(disttriang(5.0))
a = np.array(dist)
a20 = np.array(dist20)
b = np.array(triang)
```

```python
#distribuci√≥n normal con 10000 samples (realizaciones = 12)
plt.hist(a, bins = 'auto')
plt.show()
```

```python
#distribuci√≥n normal con 10000 samples (realizaciones = 20)
plt.hist(a20, bins = 'auto')
plt.show()
```

```python
#distribuci√≥n triangular con 10000 samples
plt.hist(b, bins = 'auto')
plt.show()
```
